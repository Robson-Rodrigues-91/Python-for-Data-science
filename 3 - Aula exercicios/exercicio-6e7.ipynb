6. Crie um novo dataframe, porém somente com as colunas: Survived, Pclass, Sex, Age, Ticket, Embarked

7. Faça um groupby, e retorne somente as idades. Quero que forme faixas.

```python
#importando biblioteca pandas
import pandas as pd

# carregado informacoes em um dataset
titanic_data = pd.read_csv('titanic.csv')

# Exibindo as informacoes originais 
print(titanic_data)

# Criando um novo dataset somente com as colunas passadas por parametro
new_titanic_data = titanic_data[["Survived","Pclass","Sex", "Age", "Ticket", "Embarked"]]

# Exibindo o novo dataset com as colunas setadas
print(new_titanic_data)

# Trocando o nome da coluna Age
titanic_data_idade = new_titanic_data.rename(columns={"Age":"Idade"})

# Exibindo o novo dataset
print(titanic_data_idade)

# Adicionando uma nova coluna
titanic_data_idade['Total'] = 1

# Exibindo o novo dataset com a nova coluna
print(titanic_data_idade)

# Usando groupbu para fazer a soma das idades 
titanic_data_idade[["Idade", "Total"]].groupby("Idade").sum()



```
